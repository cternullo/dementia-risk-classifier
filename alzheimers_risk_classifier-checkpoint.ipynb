# Binary Cognitive Decline Risk Classifier

This notebook builds a binary classifier to predict dementia risk using the **OASIS Longitudinal Dementia Dataset**.  
The target variable is the **Clinical Dementia Rating (CDR)**:  
- `0` → **No Dementia**  
- `> 0` → **Dementia** (mild, moderate, or severe)

We train two models — Logistic Regression and Random Forest — compare their performance, and expose the best model via a Gradio web interface.

## 1. Imports & Setup

```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import joblib

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (
    classification_report, confusion_matrix,
    roc_curve, auc, ConfusionMatrixDisplay
)

sns.set_theme(style='whitegrid', palette='muted')
plt.rcParams['figure.dpi'] = 110

RANDOM_STATE = 42
```

## 2. Load Data

We load the OASIS longitudinal CSV and take an initial look at its shape, columns, and data types.

```python
df = pd.read_csv('archive/oasis_longitudinal.csv')

print(f'Dataset shape: {df.shape}')
print(f'Columns: {df.columns.tolist()}')
df.head()
```

## 3. Exploratory Data Analysis (EDA)

Before modelling we inspect:
- **Class distribution** — how balanced is our target?
- **Missing values** — which features need imputation?
- **Feature distributions** — are there obvious outliers?
- **Correlations** — which features move together or with the target?

```python
# --- 3.1 Class distribution (raw CDR values) ---
print('CDR value counts:')
print(df['CDR'].value_counts().sort_index())

fig, ax = plt.subplots(figsize=(6, 4))
df['CDR'].value_counts().sort_index().plot(
    kind='bar', ax=ax, color=['steelblue', 'coral', 'mediumseagreen', 'tomato'], edgecolor='white'
)
ax.set_title('CDR Score Distribution (raw)', fontsize=13)
ax.set_xlabel('CDR Score')
ax.set_ylabel('Count')
ax.set_xticklabels(ax.get_xticklabels(), rotation=0)
plt.tight_layout()
plt.show()
```

```python
# --- 3.2 Binary target distribution ---
binary_counts = (df['CDR'] > 0).value_counts().rename({False: 'No Dementia (CDR=0)', True: 'Dementia (CDR>0)'})
print('Binary target distribution:')
print(binary_counts)

fig, ax = plt.subplots(figsize=(5, 4))
binary_counts.plot(kind='bar', ax=ax, color=['steelblue', 'coral'], edgecolor='white')
ax.set_title('Binary Target Distribution', fontsize=13)
ax.set_ylabel('Count')
ax.set_xticklabels(ax.get_xticklabels(), rotation=15, ha='right')
plt.tight_layout()
plt.show()
```

```python
# --- 3.3 Missing values ---
missing = df.isnull().sum()
print('Missing values per column:')
print(missing[missing > 0])

fig, ax = plt.subplots(figsize=(6, 3))
missing[missing > 0].plot(kind='barh', ax=ax, color='salmon', edgecolor='white')
ax.set_title('Columns with Missing Values', fontsize=13)
ax.set_xlabel('Missing Count')
plt.tight_layout()
plt.show()
```

```python
# --- 3.4 Feature distributions for numeric columns ---
numeric_cols = ['Age', 'EDUC', 'SES', 'MMSE', 'eTIV', 'nWBV', 'ASF']

fig, axes = plt.subplots(2, 4, figsize=(16, 7))
axes = axes.flatten()

for i, col in enumerate(numeric_cols):
    axes[i].hist(df[col].dropna(), bins=20, color='steelblue', edgecolor='white', alpha=0.85)
    axes[i].set_title(col, fontsize=11)
    axes[i].set_ylabel('Count')

# Hide the unused 8th subplot
axes[-1].set_visible(False)
fig.suptitle('Feature Distributions', fontsize=14, y=1.01)
plt.tight_layout()
plt.show()
```

```python
# --- 3.5 Correlation heatmap ---
# We temporarily encode M/F and binarise CDR just for this heatmap
df_corr = df.copy()
df_corr['Gender'] = df_corr['M/F'].map({'M': 0, 'F': 1})
df_corr['DementiaLabel'] = (df_corr['CDR'] > 0).astype(int)

corr_cols = ['Age', 'EDUC', 'SES', 'MMSE', 'eTIV', 'nWBV', 'ASF', 'Gender', 'DementiaLabel']
corr_matrix = df_corr[corr_cols].corr()

fig, ax = plt.subplots(figsize=(9, 7))
sns.heatmap(
    corr_matrix, annot=True, fmt='.2f', cmap='coolwarm',
    center=0, linewidths=0.5, ax=ax
)
ax.set_title('Feature Correlation Heatmap', fontsize=14)
plt.tight_layout()
plt.show()
```

**EDA Takeaways:**
- The dataset has moderate class imbalance — we'll use `class_weight='balanced'` to compensate.
- `SES` has 19 missing values and `MMSE` has 2; we'll impute both with their medians.
- `MMSE` has the strongest negative correlation with dementia (lower cognition score = higher risk), while `nWBV` (normalised whole brain volume) also correlates meaningfully.

## 4. Preprocessing

Steps:
1. Drop administrative / irrelevant columns.
2. Binarise the CDR target.
3. Median-impute `SES` and `MMSE`.
4. Encode `M/F` as 0/1.
5. Scale all numeric features with `StandardScaler`.

```python
# --- 4.1 Drop columns not used for modelling ---
DROP_COLS = ['Subject ID', 'MRI ID', 'Hand', 'Visit', 'MR Delay', 'Group']
df_clean = df.drop(columns=DROP_COLS).copy()

# --- 4.2 Binarise target: 0 = No Dementia, 1 = Dementia ---
df_clean['CDR'] = (df_clean['CDR'] > 0).astype(int)
df_clean.rename(columns={'CDR': 'Dementia'}, inplace=True)

# --- 4.3 Median imputation for SES and MMSE ---
ses_median = df_clean['SES'].median()
mmse_median = df_clean['MMSE'].median()
df_clean['SES'].fillna(ses_median, inplace=True)
df_clean['MMSE'].fillna(mmse_median, inplace=True)

print(f'SES median imputed: {ses_median}')
print(f'MMSE median imputed: {mmse_median}')
print(f'Remaining missing values: {df_clean.isnull().sum().sum()}')

# --- 4.4 Encode gender ---
df_clean['M/F'] = df_clean['M/F'].map({'M': 0, 'F': 1})

print('\nCleaned dataframe shape:', df_clean.shape)
df_clean.head()
```

```python
# --- 4.5 Train / test split (stratified 80/20) ---
FEATURE_COLS = ['M/F', 'Age', 'EDUC', 'SES', 'MMSE', 'eTIV', 'nWBV', 'ASF']
TARGET_COL = 'Dementia'

X = df_clean[FEATURE_COLS]
y = df_clean[TARGET_COL]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=RANDOM_STATE, stratify=y
)

print(f'Training set: {X_train.shape[0]} samples')
print(f'Test set:     {X_test.shape[0]} samples')
print(f'Train class distribution:\n{y_train.value_counts()}')
print(f'Test class distribution:\n{y_test.value_counts()}')
```

```python
# --- 4.6 Feature scaling ---
# StandardScaler fits ONLY on training data to prevent data leakage.
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled  = scaler.transform(X_test)

# Save the scaler for the Gradio app
joblib.dump(scaler, 'scaler.pkl')
print('scaler.pkl saved.')
```

## 5. Model Training

We train two classifiers:
- **Logistic Regression** — a simple, highly interpretable linear baseline.
- **Random Forest** — a non-linear ensemble that typically handles feature interactions well.

Both use `class_weight='balanced'` to compensate for the class imbalance without discarding any samples.

```python
# --- Logistic Regression ---
lr_model = LogisticRegression(
    max_iter=1000,
    class_weight='balanced',
    random_state=RANDOM_STATE
)
lr_model.fit(X_train_scaled, y_train)
print('Logistic Regression trained.')

# --- Random Forest ---
rf_model = RandomForestClassifier(
    n_estimators=300,
    max_depth=None,
    class_weight='balanced',
    random_state=RANDOM_STATE,
    n_jobs=-1
)
rf_model.fit(X_train_scaled, y_train)
print('Random Forest trained.')

# Save Random Forest for Gradio
joblib.dump(rf_model, 'rf_model.pkl')
print('rf_model.pkl saved.')
```

## 6. Evaluation

We evaluate each model with:
- **Classification report** — precision, recall, and F1 per class.
- **Confusion matrix** — visualise true vs. predicted labels.
- **ROC curve & AUC** — probability-threshold-agnostic performance measure.
- **Feature importance / coefficients** — understand which inputs drive predictions.

```python
# Helper: print classification report and plot confusion matrix
def evaluate_model(model, X_scaled, y_true, model_name, ax):
    y_pred = model.predict(X_scaled)
    print(f'\n=== {model_name} — Classification Report ===')
    print(classification_report(y_true, y_pred, target_names=['No Dementia', 'Dementia']))

    cm = confusion_matrix(y_true, y_pred)
    disp = ConfusionMatrixDisplay(cm, display_labels=['No Dementia', 'Dementia'])
    disp.plot(ax=ax, colorbar=False, cmap='Blues')
    ax.set_title(f'{model_name}\nConfusion Matrix', fontsize=12)


fig, axes = plt.subplots(1, 2, figsize=(12, 4))
evaluate_model(lr_model, X_test_scaled, y_test, 'Logistic Regression', axes[0])
evaluate_model(rf_model, X_test_scaled, y_test, 'Random Forest',       axes[1])
plt.tight_layout()
plt.show()
```

```python
# --- ROC Curves ---
fig, ax = plt.subplots(figsize=(7, 5))

for model, name, color in [
    (lr_model, 'Logistic Regression', 'steelblue'),
    (rf_model, 'Random Forest',       'coral')
]:
    y_prob = model.predict_proba(X_test_scaled)[:, 1]
    fpr, tpr, _ = roc_curve(y_test, y_prob)
    roc_auc = auc(fpr, tpr)
    ax.plot(fpr, tpr, color=color, lw=2, label=f'{name} (AUC = {roc_auc:.3f})')

ax.plot([0, 1], [0, 1], 'k--', lw=1, label='Random Classifier')
ax.set_xlabel('False Positive Rate')
ax.set_ylabel('True Positive Rate')
ax.set_title('ROC Curves', fontsize=13)
ax.legend(loc='lower right')
plt.tight_layout()
plt.show()
```

```python
# --- Feature Importance — Random Forest ---
importances = rf_model.feature_importances_
feat_df = pd.Series(importances, index=FEATURE_COLS).sort_values(ascending=True)

fig, ax = plt.subplots(figsize=(7, 5))
feat_df.plot(kind='barh', ax=ax, color='coral', edgecolor='white')
ax.set_title('Random Forest — Feature Importances', fontsize=13)
ax.set_xlabel('Mean Decrease in Impurity')
plt.tight_layout()
plt.show()
```

```python
# --- Coefficients — Logistic Regression ---
coefs = pd.Series(lr_model.coef_[0], index=FEATURE_COLS).sort_values(ascending=True)

colors = ['coral' if c > 0 else 'steelblue' for c in coefs]
fig, ax = plt.subplots(figsize=(7, 5))
coefs.plot(kind='barh', ax=ax, color=colors, edgecolor='white')
ax.axvline(0, color='black', linewidth=0.8)
ax.set_title('Logistic Regression — Coefficients\n(coral = increases risk, blue = decreases risk)', fontsize=12)
ax.set_xlabel('Coefficient Value')
plt.tight_layout()
plt.show()
```

**Evaluation Summary:**
- Both models beat random chance substantially (AUC well above 0.5).
- `MMSE` (Mini-Mental State Examination score) is consistently the most important predictor — low scores strongly indicate dementia.
- `nWBV` (normalised whole brain volume) and `Age` also rank highly.
- The Random Forest's non-linear capacity generally gives it a slight edge on AUC.
- Balanced class weights ensure recall for the minority Dementia class remains high.

## 7. Prediction Utility Function

`predict_risk()` takes a dictionary of raw (unscaled) patient values, applies the same preprocessing pipeline, and returns a dementia probability and a human-readable risk label.

```python
def predict_risk(gender: str, age: float, educ: float, ses: float,
                 mmse: float, etiv: float, nwbv: float, asf: float,
                 model=rf_model, fitted_scaler=scaler) -> dict:
    """
    Predict cognitive decline risk from raw input values.

    Parameters
    ----------
    gender : 'M' or 'F'
    age    : patient age in years
    educ   : years of education
    ses    : socioeconomic status (1=highest, 5=lowest)
    mmse   : Mini-Mental State Examination score (0–30)
    etiv   : estimated total intracranial volume (mm³)
    nwbv   : normalised whole brain volume
    asf    : atlas scaling factor

    Returns
    -------
    dict with 'probability' (float) and 'risk_label' (str)
    """
    gender_encoded = 0 if str(gender).strip().upper() == 'M' else 1
    raw = np.array([[gender_encoded, age, educ, ses, mmse, etiv, nwbv, asf]])
    raw_scaled = fitted_scaler.transform(raw)
    prob = model.predict_proba(raw_scaled)[0, 1]
    label = 'High Risk' if prob >= 0.5 else 'Low Risk'
    return {'probability': round(float(prob), 4), 'risk_label': label}


# --- Quick sanity check ---
result = predict_risk(
    gender='M', age=75, educ=12, ses=2,
    mmse=23, etiv=1678, nwbv=0.736, asf=1.046
)
print('Sample prediction:', result)
```
